## Pandas 

pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

```python
import pandas as pd
```

```python
df = pd.read_csv("Glasgow_Weather.csv")
```

```python
df
```

You can find the dimension of your data base by using the attribute shape. Be aware it is an attribute because we do not use the rounded brackets at the end.

```python
df.shape
```

The function info will give you greater detail on each column, stating number of non-null data points and the type of data.

If you see object, it normally means it is a string but if you aren't sure take a look at the data yourself.

```python
df.info()
```

To view the data you can use the function head or tail. If you leave the brackets empty it shows only 5 rows but you can tell it how many to show. 

Be aware after 60 rows it will squash the rows together just like when you output the dataframe itself.

```python
df.head()
```

```python
df.tail()
```

More indepth look at data can be done by using describe and correlation.

```python
df.describe()
```

```python
df.corr()
```

You can filter on each column on the dataframe and a single column will be called a series.

Use the dataframe name and square brackets is how to filter your data.

```python
df['tempMin']
```

Again the above will look familiar to you, as its almost like a dictionary.

If you want to view multiple columns at once, then replace the string with a list of the column names.

```python
df[ ['tempMin', 'tempMax'] ]
```

The order in which you type the list is the order in which the filtered dataframe will output.

```python
df[['tempMax', 'tempMin']]
```

To find out all the columns use the attribute columns. This is useful if you are working on a very large dataframe where the column names are squashed together when output.

```python
df.columns
```

If you are wanting to find the data for a particular row then you will use the iloc function. 

Image the iloc function like the list indexing. If you want the first row you pass 0 into the function.

```python
df.iloc[0]
```

If you want a single row, it provides a data series.

Just like if you want to slice a list use colon between the numbers in the iloc function. It will return a dataframe.

```python
df.iloc[0:5]
```

You can also pass a list of indexs in any order.

```python
df.iloc[[0,2,4,3,5]]
```

If you only want certain rows with certain columns then seperate the row definition with a comma and use the syntax as the rows.

You are passing 2 lists so remember that iloc will look like a nested list (extra square brackets). Additonally if you use colon notation, you will need to remove the brackets for the list.

```python
df.iloc[[0,1],[0,1,2]]
```

```python
df.iloc[0:2,0:3]
```

If your dataframe has an index you can still use iloc but you will find it difficult to find the index you are looking for. 

Pandas has the function loc[] which you pass in the unique index which will not be a integer to end of data rows.

In our current dataframe the index is numbers so we can still use loc with numbers.

```python
df.loc[[0,1],["summary", "cloudCover"]]
```

It makes most sense that the dataframe index would be the date, as an index should be unique, so we shall set the dataframes index to be "day"

```python
df.set_index("day")
```

```python
df
```

As you can see above the index still remain as the numbers from 0 to 1794. Pandas allows you to see what the result of your code is before it saves it. 

To save it you can either pass a paramenter of True to inplace or use assignment expression.

```python
df.set_index("day", inplace=True)
```

```python
df
```

To reset the index simply use the reset_index and pass true to inplace to save the result. Again you can also use assignment to save the result.

```python
df.reset_index(inplace=True)
df
```

```python
df = df.set_index("day")
df
```

Now you can see using loc with the unique index of day, returns the result for that row.

```python
df.loc["2019-11-30"]
```

The functionality of loc isthe same as iloc, so you can slice both rows and columns. Using list and colons.

```python
df.loc[["2019-11-30"],"summary":"cloudCover"]
```

```python
df.loc["2019-11-01":"2019-11-30","tempMax":"visibility"]
```
